```python
import numpy as np
from matplotlib import pyplot as plt
from digicomm import derivativeFilter

Tsamp = 0.01
Tstart = -2*np.pi
Tstop = 2*np.pi
t = np.arange(Tstart,Tstop,step=Tsamp)
N = 4*8-1
x = np.cos(t)
y = derivativeFilter(x,N=N,Tsamp=Tsamp)

plt.plot(t,x)
plt.plot(t,y)
plt.grid(True)
plt.legend([
    "Original Signal",
    "Derivative"
])
plt.show()

```
