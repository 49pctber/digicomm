```python
import numpy as np
import matplotlib.pyplot as plt
from digicomm import fractionalDelayFilter
```

```python
T = 0.01 # sample time
dT = T*0.5 # delay
L = 10
t = np.arange(-5,5,T)
tt = t + dT

foo = lambda t : np.cos(2*np.pi*t/2)

y = foo(t)
ya = foo(tt)
yd = fractionalDelayFilter(y, 0.5)

plt.figure(figsize=(10,10))

plt.subplot(2,1,1)
plt.plot(t,y)
plt.stem(t,y,markerfmt=' ')
plt.stem(tt,ya,markerfmt=' ')
#plt.scatter(tt,yd)
plt.grid()
plt.show()

plt.subplot(2,1,2)
plt.stem(t-dT,np.abs(yd-ya),'k',markerfmt=' ')
plt.grid()
plt.show()
```
